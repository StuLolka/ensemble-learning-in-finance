# Comparison of Ensemble Methods (No Hyperparameter Tuning)
In this section, we compare several ensemble methods —  
Random Forest, XGBoost, and Stacking — using the same data setup.

## Dataset Preparation and Train/Test Split
Create target feature (`tomorrow_open`) and split data  
chronologically into training and testing subsets.


```python
SNGSP_data = data[data['secid'] == 'SNGSP'].copy()
SNGSP_data['tomorrow_open'] = SNGSP_data['open'].shift(-1)
SNGSP_data = SNGSP_data.dropna()

X = SNGSP_data[['open', 'close', 'high', 'low', 'volume']]
y = SNGSP_data['tomorrow_open']

split_index = int(len(X) * 0.70)

X_train, X_test = X.iloc[:split_index], X.iloc[split_index:]
y_train, y_test = y.iloc[:split_index], y.iloc[split_index:]
```

## Random Forest Regressor
Demonstration of the **Bagging** approach using Random Forest.  
Evaluate model performance with MAE and R² metrics.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score

rfr_model = RandomForestRegressor(
        random_state=44,
        n_jobs=-1  
)
rfr_model.fit(X_train, y_train)

y_pred = rfr_model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"MAE: {mae:.2f}")
print(f"R²: {r2:.3f}")
```

## XGBRegressor
Demonstration of the **Boosting** approach using XGBoost.  
Compare results to Random Forest.

```python
from xgboost import XGBRegressor

xgb_model = XGBRegressor(
    random_state=44,
    n_jobs=-1  
)

xgb_model.fit(X_train, y_train)

y_pred = xgb_model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"MAE: {mae:.2f}")
print(f"R²: {r2:.3f}")
```

## Voting
Demonstration of the Voting ensemble method, where predictions from multiple base models are combined by averaging (for regression) to stabilize and improve overall performance.

```python
from sklearn.ensemble import VotingRegressor

voting_model = VotingRegressor([
    ('random_forest', rfr_model),
    ('xgboost', xgb_model)
])

voting_model.fit(X_train, y_train)
y_pred = voting_model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"MAE: {mae:.2f}")
print(f"R²: {r2:.3f}")
```

## Blending
Demonstration of the Blending method, a variant of stacking where the meta-model is trained on predictions of base models on a hold-out part of the training set. This approach allows combining strengths of different models without using cross-validation.

```python
from sklearn.linear_model import Ridge
import numpy as np

blend_split_idx = int(0.8 * len(X_train))


X_base = X_train[:blend_split_idx]
y_base = y_train[:blend_split_idx]
X_meta = X_train[blend_split_idx:]
y_meta = y_train[blend_split_idx:]

rfr_model.fit(X_base, y_base)
xgb_model.fit(X_base, y_base)

meta_features = np.column_stack([
    rfr_model.predict(X_meta),
    xgb_model.predict(X_meta)
])

meta_model = Ridge(alpha=1.0)
meta_model.fit(meta_features, y_meta)

test_meta_features = np.column_stack([
    rfr_model.predict(X_test),
    xgb_model.predict(X_test)
])

y_pred = meta_model.predict(test_meta_features)

mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"MAE: {mae:.2f}")
print(f"R²: {r2:.3f}")
```
